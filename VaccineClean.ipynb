```python
#imports
import os
import pandas as pd
import numpy as np
```

```python
#get data
mydir = os.getcwd()
fulldata = pd.read_csv(os.path.join(mydir,'VaccineData.csv'))

mydir = os.getcwd()
statesList = pd.read_csv(os.path.join(mydir,'StatesList.csv'))
```

```python
#Create blank array
cleandata = np.zeros([(len(fulldata.columns)-1)*len(statesList),6])
cleandatadf = pd.DataFrame(cleandata)
```

```python
k = 0
h = 0
for i in range(len(fulldata)):
    if k<51 and fulldata["Place"][i] == statesList["State"][k]:
        for j in range(1,len(fulldata.columns)):
            cleandatadf[1][h] = fulldata["Place"][i]
            cleandatadf[5][h] = fulldata[fulldata.columns[j]][i]
            if j%3 == 0:
                cleandatadf[2][h] = False
                cleandatadf[3][h] = False
                cleandatadf[4][h] = True
            elif (j+1)%3 == 0:
                cleandatadf[2][h] = False
                cleandatadf[3][h] = True
                cleandatadf[4][h] = False
            elif (j+2)%3 == 0:
                cleandatadf[2][h] = True
                cleandatadf[3][h] = False
                cleandatadf[4][h] = False
            h+=1
        k+=1
```

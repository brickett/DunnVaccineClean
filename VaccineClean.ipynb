```python
#imports
import os
import pandas as pd
import numpy as np
```

```python
#get data
mydir = os.getcwd()
fulldata = pd.read_csv(os.path.join(mydir,'VaccineData.csv'))

mydir = os.getcwd()
statesList = pd.read_csv(os.path.join(mydir,'StatesList.csv'))
```

```python
#Create blank array
numyears = ((len(fulldata.columns)-1)*len(statesList))//3
cleandata = np.zeros([numyears,5])
cleandatadf = pd.DataFrame(cleandata)
cleandatadf[1].astype(str)
```

```python
k = 0
h = 0
for i in range(len(fulldata)):
    if k<len(statesList) and fulldata["Place"][i] == statesList["State"][k]: #Determine that the place is a state
        for j in range(1,len(fulldata.columns)):
            cleandatadf.loc[h,1] = fulldata["Place"][i]
            if j%3 == 0:
                cleandatadf.at[h,4] = fulldata[fulldata.columns[j]][i]
                cleandatadf.at[h,0] = fulldata.columns[j-2]
                h+=1
            elif (j+1)%3 == 0:
                cleandatadf.at[h,3] = fulldata[fulldata.columns[j]][i]
                cleandatadf.at[h,0] = fulldata.columns[j-1]
            elif (j+2)%3 == 0:
                cleandatadf.at[h,2] = fulldata[fulldata.columns[j]][i]
                cleandatadf.at[h,0] = fulldata.columns[j]
        k+=1
```

```python
#imports
import os
import pandas as pd
import numpy as np
```

```python
#get data
mydir = os.getcwd()
fulldata = pd.read_csv(os.path.join(mydir,'VaccineData.csv'))

mydir = os.getcwd()
statesList = pd.read_csv(os.path.join(mydir,'StatesList.csv'))
```

```python
#Create blank array
cleandata = np.zeros([(len(fulldata.columns)-1)*len(statesList),6])
cleandatadf = pd.DataFrame(cleandata)
cleandatadf[1].astype(str)
```

```python
k = 0
h = 0
for i in range(len(fulldata)):
    if k<len(statesList) and fulldata["Place"][i] == statesList["State"][k]:
        for j in range(1,len(fulldata.columns)):
            cleandatadf[1][h] = fulldata["Place"][i]
            cleandatadf.set_value(h,5,fulldata[fulldata.columns[j]][i])
            if j%3 == 0:
                cleandatadf.set_value(h,2,False)
                cleandatadf.set_value(h,3,False)
                cleandatadf.set_value(h,4,True)
                cleandatadf.set_value(h,0,fulldata.columns[j-2])
            elif (j+1)%3 == 0:
                cleandatadf.set_value(h,2,False)
                cleandatadf.set_value(h,3,True)
                cleandatadf.set_value(h,4,False)
                cleandatadf.set_value(h,0,fulldata.columns[j-1])
            elif (j+2)%3 == 0:
                cleandatadf.set_value(h,2,True)
                cleandatadf.set_value(h,3,False)
                cleandatadf.set_value(h,4,False)
                cleandatadf.set_value(h,0,fulldata.columns[j])
            h+=1
        k+=1
```

```python
cleandatadf.columns = ['Year','State','Mean','LL','UL', 'Percent Vaccinated']
```

```python
#output
cleandatadf.to_csv(os.path.join(mydir,'VaccineDataClean.csv'))
```
